```python
#Question1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
'''Answer1.'''
from bokeh.plotting import figure, output_file, show

# Generate x and y values for the sine wave
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Create the Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line to the plot
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

```python
#Question2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python
'''Answer2.'''
from bokeh.models import ColumnDataSource

# Generate random data
N = 200
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(10, 50, N)
colors = ["#%02x%02x%02x" % (r, g, b) for r, g, b in zip(np.random.randint(0, 256, N), np.random.randint(0, 256, N), np.random.randint(0, 256, N))]

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create the Bokeh figure
p = figure(title="Scatter Plot with Random Data", x_axis_label='x', y_axis_label='y')

# Add scatter plot
p.scatter('x', 'y', size='sizes', color='colors', source=source)

# Show the plot
show(p)
```

```python
#Question3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
'''Answer3.'''
# Create the data source
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data={'fruits': fruits, 'counts': counts})

# Create the figure
p = figure(title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')

# Add the bars
p.vbar(x='fruits', top='counts', width=0.9, source=source)

# Show the plot
show(p)
```

```python
'''Answer4.'''
# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data={'hist': list(hist), 'edges': list(edges[:-1])})

# Create the figure
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add the bars
p.quad(left='edges', right='edges', top='hist', bottom=0, source=source)

# Show the plot
show(p)
```

```python
'''Answer5.'''
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar

# Generate random data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': xx.flatten(), 'y': yy.flatten(), 'z': data_heatmap.flatten()})

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')

# Add the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, fill_color='color', source=source)

# Set the color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create the color bar
colorbar = ColorBar(color_mapper=color_mapper, label_standoff=8, location='right')

# Add the color bar to the plot
p.add_layout(colorbar)



# Show the plot
show(p)
```
